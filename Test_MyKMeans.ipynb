```python
from sklearn.datasets.samples_generator import make_blobs
X, cluster_assignments = make_blobs(n_samples=200, centers=4, cluster_std=0.60, random_state=0)
```

```python
X
```

```python
cluster_assignments
```

```python
from kmeans import KMeans as MyKMeans

mykmeans = MyKMeans(k=4, max_iterations=50)
(groups, centroids) = mykmeans.fit(X.tolist())
f = open("./Kmeans_fit.txt",'w')
f.write(str(groups))
f.write("\n")
f.write(str(centroids))
f.close()
```

```python
(groups, centroids) = mykmeans.fit_extended(X.tolist())
f = open("./Kmeans_fit_extended.txt",'w') 
f.write(str(groups))
f.write("\n")
f.write(str(centroids))
f.close()
```
